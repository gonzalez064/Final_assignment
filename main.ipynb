```python
from src.analysis import load_data, prepare_data, compute_correlation
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
life_path = "data/life_expectancy.csv"
income_path = "data/income.csv"

life_df, income_df = load_data(life_path, income_path)

print("Life expectancy data:")
display(life_df.head())

print("\nIncome data:")
display(income_df.head())
```

```python
df = prepare_data(life_df, income_df)
df.head()
```

```python
correlation = compute_correlation(df)
print(f"Correlation between income and life expectancy at 65: {correlation}")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Income", y="LifeExpectancy")

plt.title("Income vs Life Expectancy at 65")
plt.xlabel("Income")
plt.ylabel("Life Expectancy at 65")
plt.tight_layout()
plt.show()
```

```python
df_nocov = df[(df['Year'] != 2020)]
print(df_nocov)
```

```python
correlation = compute_correlation(df_nocov)
print(f"Correlation between income and life expectancy at 65 (except 2020): {correlation}")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_nocov, x="Income", y="LifeExpectancy")

plt.title("Income vs Life Expectancy at 65")
plt.xlabel("Income")
plt.ylabel("Life Expectancy at 65")
plt.tight_layout()
plt.show()
```
